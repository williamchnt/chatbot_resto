```python
import pandas as pd
import ast


df = pd.read_csv('C:/Users/zmoha/OneDrive - Efrei/chatbot/Data_All_City.csv')
df
```

```python
df = df.drop(['id', 'alias', 'is_closed', 'transactions', 'phone', 'distance', 'image_url', 'coordinates', 'url'], axis=1)
df
```

```python
# créer une fonction lambda qui extrait la valeur de la première clé de chaque dictionnaire

df['location'] = df['location'].apply(ast.literal_eval)

extract_first_or_second_value = lambda x: x['address2'] if x['address1'] is None else x['address1']
# appliquer la fonction à la colonne contenant des dictionnaires
df['adresse'] = df['location'].apply(extract_first_or_second_value)
del df['location']
# afficher le dataframe résultant
df
```

```python
# créer une fonction lambda pour extraire la deuxième valeur du premier dictionnaire
df['categories'] = df['categories'].apply(ast.literal_eval)

df['categories'] = df['categories'].apply(lambda x: [{'alias': 'Unknown', 'title': 'Unknown'}] if len(x) == 0 else x)
# extract_value = lambda x: x[0].get('title', 'Unknown') if len(x) == 1 else \
#                           (x[0]['title'] if x[0]['title'] != 'Restaurants' and len(x[1]) > 1 else \
#                             (x[1].get(list(x[1].keys())[1], 'Unknown') if len(x) > 1 else x[0].get('alias', 'Unknown')))
df['Specialite'] = df['categories'].apply(lambda x: [dico['title'] for dico in x])
del df['categories']
# df['specialite'] = df['categories'].apply(extract_value)
```

```python
City = pd.read_csv("C:/Users/zmoha/OneDrive - Efrei/chatbot/City.csv")


Liste_Des_Villes = City['City'].tolist()
Liste_Des_Villes
```

```python
# Remplacez <votre clé API Yelp Fusion> par votre clé API Yelp Fusion
api_key = "OIfF_jOn9ch9BETldtCbleBeLbRgLDivlSjYdt0YoBgMUWGCytOemBoAdmPBlvVqojb6-XCfT5DaSf0A4OAEgOJNcJZ_eXE7WLNe3FCrx9EmtKtaZs9bXzmyfU4DZHYx"

# L'URL de base de l'API Yelp Fusion
url = "https://api.yelp.com/v3/businesses/search"

# Les paramètres de la requête
params = {
    "location": "",  # la ville sera remplacée dans la boucle
    "categories": "restaurants",
    "limit": 50,
    "offset": 0
}

# Les en-têtes de la requête
headers = {
    "Authorization": "Bearer " + api_key
}

# La liste des villes à parcourir
# villes = ['Abbaretz',
#  'Abbeville',
#  'Ablis',
#  'Ablon-sur-Seine',
#  'Abondant',
#  'Abrest'
#  ]

# La liste qui stockera les données des entreprises pour chaque ville
entreprises = []

# Boucle pour parcourir chaque ville
for ville in Liste_Des_Villes:
    # Modifier la ville dans les paramètres de la requête
    params["location"] = ville

    # Effectuer une requête GET à l'API Yelp Fusion
    response = requests.get(url, params=params, headers=headers)

    # Analyser la réponse JSON
    data = json.loads(response.text)
    
    # Récupérer les données des entreprises
    try:
        entreprises_ville = data["businesses"]
    except KeyError:
        print(f"La ville {ville} n'a pas pu être traitée. La clé 'businesses' n'a pas été trouvée dans la réponse JSON.")
        continue
    
    # Ajouter la ville aux données de chaque entreprise
    for entreprise in entreprises_ville:
        entreprise["ville"] = ville
    
    # Ajouter les données des entreprises de cette ville à la liste globale
    entreprises.extend(entreprises_ville)
    
# Convertir la liste des entreprises en un dataframe pandas
df = pd.DataFrame(entreprises)
```
